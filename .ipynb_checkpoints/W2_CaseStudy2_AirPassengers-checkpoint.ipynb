```python
import pandas as pd
from pandas import Series, DataFrame
df_time_series = pd.read_csv("AirPassengers.csv")
df_time_series.head()
```

```python
df_time_series["step"] = range(len(df_time_series))
df_time_series.head()
```

```python
df_time_series["cum_sum"] = df_time_series["#Passengers"].cumsum()
df_time_series["cum_max"] = df_time_series["#Passengers"].cummax()
df_time_series["cum_min"] = df_time_series["#Passengers"].cummin()
df_time_series.head()
```

```python
import numpy as np
temp_date = df_time_series["Month"].map(lambda x : x.split("-"))
temp_date = np.array(temp_date.values.tolist())

temp_df=DataFrame(columns=["Year", "Month"])
temp_df["Year"] = temp_date[:, 0]
temp_df["Month"] = temp_date[:, 1]
temp_df["#Passengers"] = df_time_series["#Passengers"]
temp_df
```

```python
del df_time_series["Month"]
df_time_series=pd.merge(temp_df,df_time_series, on="#Passengers")
df_time_series
```

```python
df_time_series["diff"] = df_time_series["#Passengers"].diff().fillna(0)
df_time_series[:5]
```

```python
df_time_series["pct_change"] = df_time_series["#Passengers"].pct_change().map(lambda x : x*100).fillna(0).map(lambda x : " %.2f" % x)
df_time_series
```
